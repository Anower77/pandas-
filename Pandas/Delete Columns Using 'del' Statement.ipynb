# **Delete Columns Using 'loc' and 'iloc' Method**

```python
df = df.loc[:, ['Total Screen Time ', 'Reading and Reference', ]] # loc[row, clo]
df.head()
```

```python
df = df.iloc[:, [1]] # iloc[row, clo]
df.head()
```

# **Delete Columns Using 'drop' Method**

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/rashakil-ds/Public-Datasets/main/Screen%20Time%20Data.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.drop('index', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=['Week Day', 'Social Networking', 'Reading and Reference'])
```

```python
df.drop(['Date', 'Other'], axis=1, inplace=True)
```

```python
df.head()
```
